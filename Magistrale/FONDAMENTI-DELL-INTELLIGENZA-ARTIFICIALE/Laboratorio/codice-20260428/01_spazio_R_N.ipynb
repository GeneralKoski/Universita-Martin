# Lo spazio metrico $\mathbf{R}^N$

Sia $N \in \mathbf{N}$, vogliamo definire un vettore $\underline{x} \in \mathbf{R}^N$ usando `numpy`

```python
import numpy as np
```

```python
N = 10

zero = np.zeros(N)
one = np.ones(N)
```

```python
print(zero)
print(one)

print(zero.dtype)
```

```python
x = np.array([2.0, 5, 1, 0, 7, 2, 3, 47, 3, 0])
y = np.array([3, 42, 12, 2, 5, 0, -2, -23, 0, 1])

print(x.shape)
print(y.shape)
```

```python
z = np.arange(N)

print(z)
```

```python
print(x.dtype)
print(y.dtype)
```

```python
s = np.linspace(0, 100, num=N)

print(s)
```

```python
print(s.shape)
print(s.dtype)
```

### Somma tra vettori:

```python
z = x + y
print(z)

print(z.dtype)
```

### Prodotto per scalari:

```python
print(3 * x + 2 * y)
```

```python
print( np.linalg.norm(x) )

print( np.sqrt( np.dot(x, x) ) )
```
