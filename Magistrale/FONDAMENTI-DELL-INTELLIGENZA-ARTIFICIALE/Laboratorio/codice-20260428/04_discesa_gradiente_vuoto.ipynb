#### Loss $L_1$, detta "geometria del taxi".
$$
L_1 : \mathbb{R}^{2N} \rightarrow \mathbb{R}
$$

$$
L_1(\underline{x}, \underline{y}) = \frac{1}{N} \sum_{i = 1}^N {| x_i - y_i |}
$$
dove $\underline{x} \in \mathbb{R}^N$ e $\underline{y} \in \mathbb{R}^N$.

$$

$$

$$

$$
dove $\underline{x} \in \mathbb{R}^N$ e $\underline{y} \in \mathbb{R}^N$.
